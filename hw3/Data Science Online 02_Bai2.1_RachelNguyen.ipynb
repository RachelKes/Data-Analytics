```python
import csv 
import matplotlib
from matplotlib import pyplot as plt
```

# 1. Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a,b)

```python
def linearFunc(x,a,b):
    y = [(a*i) + b for i in x]  
    plt.plot(x,y)
    return plt.show()
```

```python
def main():    
    x = [float(i) for i in input("Enter a list of x values (more than 1 value) separated by space: ").split()]
    a = float(input("Enter a number value for slope: "))
    b = float(input("Enter a number value for intercept: "))
    linearFunc(x, a, b)

if __name__ == "__main__":
    main()
```

# 2. Đọc và Vẽ biểu đồ dữ liệu chứng khoán

```python
def openFileRead(csvFile):
    indexList = []
    with open (csvFile, "r") as f:
        dataReader = csv.reader(f)
        next(dataReader) # skip the first row
        for i,row in enumerate(dataReader):
            indexList.append(row)
        return indexList
```

```python
def loadData(indexList):
    closeList = []
    openList = []
    highList = []
    for row in indexList:
        closeList.append(float(row[10]))
        openList.append(float(row[7]))
        highList.append(float(row[8]))
        
    fig = plt.figure(figsize = (10,4))
    axes = fig.add_subplot(1,1,1)
    axes.plot(closeList, color = "red", label = "close")
    axes.plot(openList, color = "green", label = "open")
    axes.plot(highList, color = "pink", label = "high")
    plt.title("Stock Prices of ABC bank")
#     plt.xlabel("Stock Price")
#     plt.ylabel("Time")
    plt.legend()
    plt.tight_layout()
    return plt.show()
```

```python
def main():
    acb_file = "excel_acb.csv"
    indexList = openFileRead(acb_file)
    loadData(indexList)         
    
if __name__ == "__main__":
    main()
```

Vẽ biểu đồ dữ liệu chứng khoán với nhiều mã

```python
def readData(csvFile):
    closeList = []
    with open (csvFile, "r") as f:
        dataReader = csv.reader(f)
        next(dataReader) # skip the first row
        for i, row in enumerate(dataReader):
            closeList.append(float(row[10]))
        return closeList
```

```python
def main():
    acb = "excel_acb.csv"
    dxg = "excel_dxg.csv"
    flc = "excel_flc.csv"
    fpt = "excel_fpt.csv"
    vjc = "excel_vjc.csv"
    vnm = "excel_vnm.csv"   
    
    fig = plt.figure(figsize = (11,5))
    axes = fig.add_subplot(1,1,1)
    axes.plot(readData(acb), "r^")
    axes.plot(readData(dxg), "go")
    axes.plot(readData(flc), "y*")
    axes.plot(readData(fpt), "m.")
    axes.plot(readData(vjc), "c")
    axes.plot(readData(vnm), "p", alpha = .2, linewidth = .5)
    plt.legend(labels = ["ACB", "DXG", "FLC", "FPT", "VJC", "VNM"], loc = [1.05, .5], frameon = False)
    plt.tight_layout()
    plt.title("The Closing Prices of Six Indexes")
    plt.gca()
    plt.show()
      

if __name__ == "__main__":
    main()
```

ve phương trình y=ax^2 + bx + c (-5 <= x <=5)

```python
def quadraticFunc():
    a = float(input("Enter the value of a: "))
    b = float(input("Enter the value of b: "))
    c = float(input("Enter the value of c: "))
    x = [i for i in range(-5, 6, 1)]
    y = [a*i**2 + b*i + c for i in x]

    fig= plt.figure()
    axes = fig.add_subplot(1,1,1)
    axes.plot(x,y, label = 'quadratic', color="red", alpha=0.5, linestyle="solid", linewidth=2)
    plt.show()

quadraticFunc()
```
