```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
np.set_printoptions(precision=3, suppress=True)
print(pd.__version__)
```

```python
path = 'supermarket_sales_vn.csv'
mkdata = pd.read_csv(path)
mkdata.head()
```

Univariate Analysis

```python
city_counts = mkdata['City'].value_counts()
city_counts
```

```python
prod_counts = mkdata['Product line'].value_counts()
prod_counts
```

```python
gender_counts = mkdata['Gender'].value_counts()
gender_counts
```

```python
payment_counts = mkdata['Payment'].value_counts()
payment_counts
```

```python
sns.set(rc={'figure.figsize':(12, 4)})
sns.set_style('darkgrid')
sns.set_palette('bright')
fig, axes = plt.subplots(1, 4)
ax1 = sns.barplot(x=city_counts.index, y=city_counts.values, orient='v', ax=axes[0])
ax1.set(title= 'City')

ax2 = sns.barplot(x=prod_counts.index, y=prod_counts.values,  orient='v' , ax=axes[1])
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=-65)
# ax2.set(xlabel = 'Product line', ylabel = 'Count')
ax2.set(title= 'Product line')

ax3 = sns.barplot(x=gender_counts.index, y=gender_counts.values, ax=axes[2])
ax3.set(title= 'Gender')

# pie chart
plt.pie(payment_counts.values, labels = payment_counts.index, autopct='%.0f%%')
plt.title('Payment')

plt.tight_layout()
```

```python
# What day is the busiest day of the week?
dayofweek = ['Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday', 'Sunday']
mkdata['dayofweek'] = pd.to_datetime(mkdata['Date']).dt.day_name()
busiest_dow = mkdata['dayofweek'].value_counts()
busiest_dow
```

```python
# What duration is the busiest time of the day?
mkdata['tod'] = pd.to_datetime(mkdata['Time']).dt.hour
tod = pd.Series(['Evening'] * mkdata['tod'].size)
tod[mkdata['tod'] < 12] = 'Morning'
tod[(12 <= mkdata['tod']) & (mkdata['tod'] <18)] = "Afternoon"
mkdata['timeofday'] = tod
mkdata.drop(['tod'], inplace=True, axis=1)
busiest_tod = mkdata['timeofday'].value_counts()
busiest_tod
```

```python
sns.set(rc={'figure.figsize':(6, 4)})
fig, ax = plt.subplots(1,2) 
ax1 = sns.barplot(x=busiest_dow.index, y=busiest_dow.values, ax=ax[0], errcolor='k')
ax1.set(xticklabels = busiest_dow.index, title='The busiest day of the week')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=-65)
# ax1.legend(busiest_dow.index, loc=[1.1, 0.5])

ax2 = sns.barplot(x=busiest_tod.index, y=busiest_tod.values, ax=ax[1])
ax2.set(xticklabels = busiest_tod.index,title='The busiest period of the day' )

plt.tight_layout()
```

Multivariate Analysis

```python
total_attr = pd.pivot_table(mkdata, index = ['City', 'Gender'], columns=['Product line'], values = 'Total')
total_attr.reset_index(inplace=True)
total_attr.head()
```

```python
city_total_sum = total_attr.groupby('City').sum()  #.loc[:,'Electronic accessories': ]
city_total_sum.reset_index(level = 0, inplace = True)
city_total_sum
```

```python
total_att = pd.pivot_table(mkdata, index = ['Product line', 'Gender'], columns=['City'], values = 'Total')
total_att.reset_index(level = 0, inplace=True)
total_hn = total_att.groupby('Product line')['Hà Nội'].sum()
total_sg = total_att.groupby('Product line')['TP HCM'].sum()
total_dn = total_att.groupby('Product line')['Đà Nẵng'].sum()
# total_att.head()
total_hn
```

```python
total_gender = pd.pivot_table(mkdata, index = ['Product line'], columns=['Gender'], values = 'Total')
total_female = total_gender.groupby('Product line')['Nữ'].sum()
total_male = total_gender.groupby('Product line')['Nam'].sum()
# total_female
# total_gender
```

```python
fig, ax = plt.subplots(figsize=(12,4))

sns.set(style="whitegrid")
sns.lineplot(data=total_hn, x=total_hn.index, y=total_hn.values, marker='o', color='red') 
sns.lineplot(data=total_sg, x=total_sg.index, y=total_sg.values, marker='v', color='blue')
sns.lineplot(data=total_dn, x=total_dn.index, y=total_dn.values, marker='s', color='green') 

labels = total_attr['City'].unique()
ax.set(title='Total by Cities')
ax.legend(labels, loc=[1.01, 0.4])

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(11,4))

sns.set(style="whitegrid")
sns.lineplot(data=total_gender, x=total_female.index, y=total_female.values, marker='o', color='red') 
sns.lineplot(data=total_gender, x=total_male.index, y=total_male.values, marker='v', color='blue')

ax.set(title='Consumption Average by Gender')
ax.legend(['Female', 'Male'], loc=[1.01, 0.4])

plt.tight_layout()
```

● Thuế: tương tự doanh thu

```python
tax_att = mkdata[['City', 'Gender', 'Product line', 'Tax 5%']]
tax_hn_gender = tax_att[tax_att['City'] == 'Hà Nội'].groupby(['Product line', 'Gender']).sum().reset_index()
tax_sg_gender = tax_att[tax_att['City'] == 'TP HCM'].groupby(['Product line', 'Gender']).sum().reset_index()
tax_dn_gender = tax_att[tax_att['City'] == 'Đà Nẵng'].groupby(['Product line', 'Gender']).sum().reset_index()

tax_hn_gender.head()
```

```python
sns.set(rc={'figure.figsize':(12, 4)})
fig, ax = plt.subplots(1,3)

sns.set(style="whitegrid")
sns.set_palette(sns.color_palette("colorblind"))

ax1 = sns.lineplot(data=tax_hn_gender, x='Product line', y='Tax 5%', hue='Gender', ax=ax[0], marker='o') 
ax2 = sns.lineplot(data=tax_sg_gender, x='Product line', y='Tax 5%', hue='Gender', ax=ax[1], marker='o') 
ax3 = sns.lineplot(data=tax_dn_gender, x='Product line', y='Tax 5%', hue='Gender', ax=ax[2], marker='o') 

ax1.set(title='Sum Tax by Gender in Hanoi')
ax1.tick_params('x', labelrotation= 45)

ax2.set(title='Sum Tax by Gender in Saigon')
ax2.tick_params('x', labelrotation= 45)

ax3.set(title='Sum Tax by Gender in Danang')
ax3.tick_params('x', labelrotation= 45)

fig.tight_layout()
```

● Rating: so sánh rating giữa các nhóm hàng, giữa các ngày trong tuần, giới tính

```python
rating_attr = mkdata[['Product line', 'dayofweek', 'Gender', 'Rating']]
rating_female_dow = rating_attr[rating_attr['Gender'] == 'Nữ'].groupby(['Product line', 'dayofweek']).mean().reset_index()
rating_male_dow = rating_attr[rating_attr['Gender'] == 'Nam'].groupby(['Product line', 'dayofweek']).mean().reset_index()

rating_female_dow.head(3)
```

```python
sns.set(rc={'figure.figsize':(12, 4)})
fig, ax = plt.subplots(1,2)

sns.set(style="white")
sns.set_palette(sns.color_palette("deep"))

ax1 = sns.lineplot(data=rating_female_dow, x='dayofweek', y='Rating', hue='Product line', ax=ax[0], marker='o', legend=False) 
ax2 = sns.lineplot(data=rating_male_dow, x='dayofweek', y='Rating', hue='Product line', ax=ax[1], marker='o', legend=False) 

ax1.set(title='Average Rating by Female')
ax1.tick_params('x', labelrotation= 45)

ax2.set(title='Average Rating by Male')
ax2.tick_params('x', labelrotation= 45)

label_vals = rating_attr['Product line'].unique()
fig.legend(title='Product line', loc=[0.45, 0.01], labels=label_vals)
fig.tight_layout()
```
