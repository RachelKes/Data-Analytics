### Phân tích đơn biến

#### Tổng đơn
 - Theo chi nhánh (City)
 - Theo nhóm hàng (Customer type)
 - Giới tính khách hàng (Gender)
 - Phương thức thanh toán (Payment)

```python
# Tổng đơn
def draw_invoice_count_dist(df, col_name, x_label, y_label, title, cur_ax):
    temp_df = df[col_name].value_counts()
    cur_ax = sns.barplot(x=temp_df.index, y=temp_df, ax=cur_ax)
    cur_ax.set_xlabel(x_label)
    cur_ax.set_ylabel(y_label)
    cur_ax.set_title(title)
    
    return cur_ax
```

```python
import math
```

```python
observed_attrs = {'City': 'Chi nhánh',
                  'Customer type': 'Nhóm khách hàng',
                  'Gender': 'Giới tính',
                  'Payment': 'Phương thức thanh toán'}

nrows = math.ceil(len(observed_attrs) / 2)
fig, axes = plt.subplots(nrows,2, figsize=(16,8), constrained_layout=True)

for i, kv in enumerate(observed_attrs.items()):
    x_label = kv[1]
    y_label = 'Số lượng đơn hàng'
    title = f'Số lượng tổng đơn theo\n{kv[1].lower()}'
    
    index = i // 2
    sub_index = i % 2
    draw_invoice_count_dist(
        df=df,
        col_name=kv[0],
        x_label=x_label,
        y_label=y_label,
        title=title,
        cur_ax=axes[index][sub_index]
    )
    
```

- `TPHCM` chiếm gần phân nửa số lượng đơn hàng so với các chi nhánh còn lại
- Các thuộc tính `Nhóm khách hàng` và `Giới tính` có phân bố số lượng đơn hàng đều nhau.
- `Credit card` có vẻ có ít số lượng đơn hàng hơn so với các phương thức còn lại

### Phân tích đa biến

#### Tổng doanh thu
1. Tỷ lệ tổng doanh thu theo chi nhánh/nhóm khách hàng/giới tính
2. Mối quan hệ giữa doanh thu (Total) với các thuộc tính như chi nhánh/nhóm khách hàng/giới tính

```python
def draw_total_dist(df, col_name_1, col_name_2, x_label, y_label, title, cur_ax):
    temp_df = df.groupby(col_name_1)[col_name_2].sum()
    cur_ax = sns.barplot(x=temp_df.index, y=temp_df, ax=cur_ax)
    cur_ax.set_xlabel(x_label)
    cur_ax.set_ylabel(y_label)
    cur_ax.set_title(title)
    
    return cur_ax
```

```python
# Tỷ lệ tổng doanh thu theo chi nhánh/nhóm khách hàng/giới tính

observed_attrs = {'City': 'Chi nhánh',
                  'Customer type': 'Nhóm khách hàng',
                  'Gender': 'Giới tính'}

nrows = math.ceil(len(observed_attrs) / 2)
fig, axes = plt.subplots(nrows,2, figsize=(16,8), constrained_layout=True)

for i, kv in enumerate(observed_attrs.items()):
    x_label = kv[1]
    y_label = 'Tổng doang thu'
    title = f'Tổng doang thu theo\n{kv[1].lower()}'
    
    index = i // 2
    sub_index = i % 2
    draw_total_dist(
        df=df,
        col_name_1=kv[0],
        col_name_2='Total',
        x_label=x_label,
        y_label=y_label,
        title=title,
        cur_ax=axes[index][sub_index]
    )
    
```

```python
# Phân bố doanh thu theo chi nhánh/nhóm khách hàng/giới tính
fig, axes = plt.subplots(1,3, figsize=(12,4), constrained_layout=True)
sns.boxplot(data=df, x='City', y='Total', ax=axes[0])
sns.boxplot(data=df, x='Customer type', y='Total', ax=axes[1])
sns.boxplot(data=df, x='Gender', y='Total', ax=axes[2])
```

```python
# Tỷ lệ tổng thuế theo chi nhánh/nhóm khách hàng/giới tính

nrows = math.ceil(len(observed_attrs) / 2)
fig, axes = plt.subplots(nrows,2, figsize=(16,8), constrained_layout=True)

for i, kv in enumerate(observed_attrs.items()):
    x_label = kv[1]
    y_label = 'Tổng thuế'
    title = f'Tổng thuế theo\n{kv[1].lower()}'
    
    index = i // 2
    sub_index = i % 2
    draw_total_dist(
        df=df,
        col_name_1=kv[0],
        col_name_2='Tax 5%',
        x_label=x_label,
        y_label=y_label,
        title=title,
        cur_ax=axes[index][sub_index]
    )
```

```python
# Phân bố thuế theo chi nhánh/nhóm khách hàng/giới tính
fig, axes = plt.subplots(1,3, figsize=(12,4), constrained_layout=True)
sns.boxplot(data=df, x='City', y='Tax 5%', ax=axes[0])
sns.boxplot(data=df, x='Customer type', y='Tax 5%', ax=axes[1])
sns.boxplot(data=df, x='Gender', y='Tax 5%', ax=axes[2])
```
