```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision=3, suppress=True)
# pd.set_option('display.float_format', lambda x: '%.3f' % x)

print(pd.__version__)
```

Read csv file

```python
filepath = 'products.csv'
product_info = pd.read_csv(filepath)
product_info.head()
```

```python
N = 5e5
filepath = 'order_products__prior.csv'
order_product_info = pd.read_csv(filepath, chunksize=N)
order_product_info_chunk = next(order_product_info)
order_product_info_chunk.head()
```

```python
filepath = 'departments.csv'
department_info = pd.read_csv(filepath)
department_info.head()
```

Order_products_prior has product_id (the order products' info) 
Products has product_id, department_id (the products' info)
Departments has department_id, department(the department' info)

```python
def selectDeptId(order_prod_id):
    # print(order_prod_id.head())
    product_dept_id = product_info[['product_id', 'department_id']]
    order_dept_id = pd.merge(order_prod_id, product_dept_id, on='product_id')['department_id']
    dept_id_count = order_dept_id.value_counts().reset_index()
    dept_id_count.columns = ['department_id', 'counts']
    return dept_id_count

chunkNum = 5e5
file = 'order_products__prior.csv'
order_products = pd.read_csv(file, chunksize=chunkNum, usecols=['product_id'])
chunk = next(order_products)
total_dept_count = selectDeptId(chunk)

for i, chunk in enumerate(order_products):
    chunk_count = selectDeptId(chunk)
    total_dept_count = pd.concat([total_dept_count, chunk_count])
    # print(total_dept_count.shape)
    
print(total_dept_count.head())
```

10 departments have the largest products ordered

```python
total_dept_count = total_dept_count.groupby('department_id').sum().reset_index()
total_dept_count.head()

# 10 departments with largest ordered products 
n_dept_id = total_dept_count.nlargest(10, columns='counts')
n_dept_id
```

Bar Plot

```python
n_dept_id_name = pd.merge(n_dept_id, department_info, on='department_id')
n_dept_id_name
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()
ax = n_dept_id_name.plot.bar(ax=ax, x='department', y='counts', rot=60, width=0.8, 
                           color='red', legend=None)
# ax.bar_label(container=ax.containers[0], padding=2, fontsize=6, rotation=0)
ax.set_xticks(n_dept_id_name.index)
ax.set_xticklabels(n_dept_id_name['department'])
plt.tight_layout()
plt.show()
plt.close('all')
```

Treemap plot

```python
import squarify

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()
labels = rest_order_prod_name['department']
size = rest_order_prod_name['counts']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

ax = squarify.plot(sizes = size, label=labels, pad=False, color=colors, alpha=.85)
plt.axis('off')
plt.show()
plt.close('all')
```
